### Find the top 10 Cuisines

```python
full_df['Cuisines']
```

```python
full_df = full_df.dropna()
```

```python
full_df['Cuisines'] = full_df['Cuisines'].apply(lambda x: x.strip())
```

```python
full_df.shape
```

```python
cuisine_list = []
for val in full_df['Cuisines']:
    cus = val.split(',')

    for i in cus:
        cuisine_list.append(i)

print(cuisine_list)
```

```python
cusine_dict = {i:cuisine_list.count(i) for i in cuisine_list}

sorted_dict = sorted(cusine_dict.items(), key=lambda x:x[1], reverse=True)
```

```python
sorted_dict = dict(sorted_dict)
sorted_dict
```

```python
keys = list(sorted_dict.keys())[:10]
vals = list(sorted_dict.values())[:10]
sns.barplot(x= keys , y= vals)
plt.xticks(rotation = 45)
plt.show()
```
