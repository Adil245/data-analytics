```python
# Data Cleaning- removing those rows where 80% values are null and replacing remaning null values with mean of the column
def data_cleaning(df, null_threshold=0.8):
   
    null_threshold = int(null_threshold * len(df.columns))
    
    numeric_cols = df.select_dtypes(include='number').columns
    
    null_counts = df.isnull().sum(axis=1)
    
    # Keep rows with non-null values less than the threshold
    df_filtered = df[null_counts < null_threshold]
    
    df_filled = df_filtered.copy()
    df_filled[numeric_cols] = df_filled[numeric_cols].apply(lambda x: x.fillna(x.mean()))

    return df_filled
```
