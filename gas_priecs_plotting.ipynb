```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gas_prices.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
plt.figure(figsize=(10,7))

for col in df.columns[1:]:
    plt.plot(df['Year'] , df[col], label =col)

plt.title('Gas price every year')
plt.xticks(df.Year, rotation=60)
plt.xlabel('Year')
plt.ylabel('Price in $')
plt.legend()
plt.show()
```
