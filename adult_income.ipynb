## Analysis of data

```python
# age
plt.figure(figsize=(8,4))
sns.histplot(df['age'], color='green',bins=18)
plt.title('AGE Distribution')
plt.show()
```

```python
sns.histplot(x=df['age'], hue=df['salary'], color='green', bins=15)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x=df['workclass'], hue=df['salary'], palette=['#010A07', '#E62A97'])
plt.title('Workclass distribution')
plt.xticks(rotation= 60)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(x=df['country'], hue=df['workclass'], color='green', bins=15)
plt.title('country wise Distribution of workclass')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.countplot(x=df['occupation'], hue=df['salary'], palette=['#010A07', '#E62A97'])
plt.title('occupation distribution')
plt.xticks(rotation= 60)
plt.show()
```

```python
#Gender
sns.countplot(y=df['gender'], hue=df['salary'], palette=['#010A07', '#E62A97'])
plt.title('occupation distribution')
plt.show()
```

```python
# capital loss
sns.histplot(x=df['capital-loss'], hue=df['salary'], bins=15)
plt.title('capital loss')
plt.show()
```

```python
sns.histplot(x=df['capital-gain'], hue=df['salary'], bins=15)
plt.title('capital gain')
plt.show()
```

```python
def hers_def(val):
    if val < 40:
        return '<40hrs'
    elif val ==40:
        return '40hrs'
    else:
        return '>40hrs'
```

```python
df['hours-per-week']= df['hours-per-week'].apply(hers_def)
```

```python
# df.head()
df['hours-per-week'].value_counts()
```

```python
sns.countplot(x=df['hours-per-week'], hue=df['salary'], palette='viridis', saturation=0.9,
              order=['<40hrs', '40hrs','>40hrs'])
plt.tight_layout()
plt.title('hours per week vs salary')
plt.show()
```
