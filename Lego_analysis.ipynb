## 1.What Percentage of all licensed sets ever released were Star Wars themed

```python
theme.loc[theme['name'] =='Star Wars']
```

```python
df['parent_theme'].value_counts()['Star Wars']
```

```python
# No. of rows
df.shape[0]
```

```python
merged_df = df.merge(theme , left_on='parent_theme', right_on='name')
```

```python
merged_df.drop(columns='name_y', inplace=True)
merged_df
```

```python
licensed = merged_df.loc[merged_df['is_licensed'] == True]
licensed
```

```python
licensed['set_num'].isnull().sum()
```

```python
licensed.dropna(subset=['set_num'], inplace=True)
```

```python
licensed['is_licensed'].unique()
```

```python
tot_rows = licensed.shape[0]
tot_rows
```

```python
# merged_df['parent_theme'].value_counts()
star_wars = merged_df['parent_theme'].value_counts()['Star Wars']
```

```python
star_wars
```

```python
answer = (star_wars/tot_rows)*100
print('Star wars theme has a percentage of ',"%.2f" % answer ,'% of all licenced themes ever released.')
```

## In which year star war was not the most popular  licenced theme (In terms of number of stes released that year)

```python
licensed.head()
```

```python
licensed['conut'] = 1
```

```python
licensed.head()
```

```python
grouped_df = licensed.groupby(['year', 'parent_theme'])['conut'].sum().reset_index()
grouped_df.sort_values('conut' , ascending=False).drop_duplicates(['year'])
```

### 2017 was the year where star wars was not the most populer licenced theme
