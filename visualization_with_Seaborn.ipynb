```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

### plotting with Seaborn

```python
res = sns.barplot(x=df['cyl'], y=df['mpg'])
plt.show()
```

```python
res= sns.histplot(x=df['mpg'], bins=10, color='g', kde=True)
plt.show()
```

As the no. of cylinder increases miles per gallons of the cars decreases

```python
res  = sns.barplot(x=df['gear'], y=df['hp'] , palette='rocket')
plt.show()
```

horse power is high for the mtcars which have 5 gears

```python
sns.countplot(x=df['carb'], palette="Set2")
```

Most of the engines are using 2 and 4 type of carburator

```python
res= sns.countplot(x=df['gear'], hue=df['cyl'], palette='mako')
plt.show()
```

```python
plt.figure(figsize=(10,8))
res= sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
df_iris.head()
```

```python
df_iris.info()
```

```python
df_iris['species'].value_counts()
```

```python
sns.scatterplot(x=df_iris['sepal_length'], y=df_iris['petal_length'], hue = df_iris['species'], 
                size=df_iris['species'])
```

```python
sns.pairplot(df_iris, hue='species')
plt.show()
```

### Petal length is the label by which we can clearly diffrentiate between the species of iris

```python
sns.lmplot(x='petal_length', y= 'petal_width', hue='species', data = df_iris, markers=['o', '*', '^'])
```

```python
sns.boxplot(x='species', y='petal_length', data=df_iris)
plt.title('stats of petal length of every species')
plt.show()
```

## Now working with the datasets of car_crashes 

```python
car_crashes = sns.load_dataset('car_crashes')
car_crashes.head()
```

```python
car_crashes.info()
```

```python
car_crashes.shape
```

```python
car_crashes.head()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='abbrev', y='total', data= car_crashes, order=car_crashes.sort_values('total').abbrev)
plt.xticks(rotation = 90)
plt.xlabel(' US States')
plt.title(' Total no. of Collision in every state')
plt.show()
```

We can clearly see that washigton dc has least no. of collision and South Carolina and North Dakota has highest collsions

```python
sns.histplot(x=car_crashes['total'], bins =10, kde=True)
plt.title('total number of collision per billin mile')
plt.show()
```

```python
sns.histplot(x=car_crashes['not_distracted'],  kde=True)
plt.title('no. of times driver was not distracted')
plt.show()
```

```python
sns.pairplot(car_crashes)
```

```python
sns.heatmap(car_crashes.corr(numeric_only=True), annot=True)
```
